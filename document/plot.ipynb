```python
from pathlib import Path

import matplotlib.pyplot as plt
import matplotlib.axes._axes
import numpy as np
import seaborn as sns

from HexLattice.plot_config import PlotConfig, high_contrast_colors
```

```python
# set plot config
pc = PlotConfig('Plot of Axial against Ring')
pc.image_root_dir = Path('document/image')
pc.plot_style =  'bmh'
pc.set_plot_config()

# prepare data
k = np.linspace(0, 6, 100)
r = 1
x = ( + abs(k - 2 * r)  \
        + abs(k - 3 * r)  \
        - abs(k - 5 * r)  \
    ) / 2    - k / 2   + r      

z = ( - abs(k - 1 * r)  \
        - abs(k - 2 * r)  \
        + abs(k - 4 * r)  \
        + abs(k - 5 * r)  \
    ) / 2               \
        + k - 3 * r
        
y = - x - z
# plot
fig= plt.figure(figsize=(8, 6))
axes = fig.subplots(1, 1, sharex=True)
variables = [x, y, z]
labels = ['x', 'y', 'z']
palette = sns.color_palette("deep")
for i, var in enumerate(variables):
    ax: matplotlib.axes._axes.Axes = axes
    ax.plot(k, var, label=f'Variable {labels[i]}', color=palette[i])
    
    # xticks
    ax.set_xticks([x for x in range(7)])
    ax.set_xticklabels([0, 'r'] + [f'{x}r' for x in range(2, 7)])
    
    # yticks
    ax.set_yticks([x for x in range(-1, 2, 1)])
    ax.set_yticklabels(['-r', 0, 'r'])
    
    ax.set_xlim((-0.2,6.2))
    ax.set_ylim((-1.2,1.2))
    
    ax.legend(loc='best')
axes.set_ylabel('Axial Coordinate')    
axes.set_xlabel('Ring Coordinate k')
plt.savefig(pc.image_path)
```

```python
# set plot config
pc = PlotConfig('Plot of Axial_all against Ring')
pc.image_root_dir = Path('document/image')
pc.plot_style =  'bmh'
pc.set_plot_config()

# prepare data
k = np.linspace(0, 6, 100)
r = 1
x = ( + abs(k - 2 * r)  \
        + abs(k - 3 * r)  \
        - abs(k - 5 * r)  \
    ) / 2    - k / 2   + r      

z = ( - abs(k - 1 * r)  \
        - abs(k - 2 * r)  \
        + abs(k - 4 * r)  \
        + abs(k - 5 * r)  \
    ) / 2               \
        + k - 3 * r
        
y = - x - z

t = ( abs(x) + abs(y) + abs(z))
# plot
fig= plt.figure(figsize=(8, 6))
axes = fig.subplots(1, 1, sharex=True)
ax: matplotlib.axes._axes.Axes = axes
ax.plot(k, t, label=f'Variable t')

# xticks
ax.set_xticks([x for x in range(7)])
ax.set_xticklabels([0, 'r'] + [f'{x}r' for x in range(2, 7)])

# yticks
ax.set_yticks([x for x in range(7)])
ax.set_yticklabels([0, 'r'] + [f'{x}r' for x in range(2, 7)])

ax.set_xlim((-0.2,6.2))
ax.set_ylim((-6.2,6.2))

ax.legend(loc='best')
axes.set_ylabel('Axial Coordinate')    
axes.set_xlabel('Ring Coordinate k')
plt.show()
```
